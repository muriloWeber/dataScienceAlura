```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
```

```python
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
```

```python
# display Ã© do prÃ³prio notebook, nÃ£o do python
display(deaths.head())
display(recovered.head())
```

```python
def latest_by_country(data):
  return data.groupby('Country/Region').sum().iloc[:,-1]

latest_by_country(confirmed)
```

```python
def latest_by_country_at(data, date):
  return data.groupby('Country/Region').sum()[date]

informations = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
informations_2 = [latest_by_country_at(confirmed, '2/20/20'), latest_by_country_at(deaths, '2/20/20'), latest_by_country_at(recovered, '2/20/20')]

combined = pd.concat(informations, axis = 1)
combined.columns = ['confirmed', 'deaths', 'recovered']
combined.head()
```

```python
# taxa de letalidade Ã© expressa em %
# o nÃºmero de casos letais dentro de todos os casos

sum_up = combined.sum()
letality_rate_1 = sum_up['deaths'] / sum_up['confirmed']
print(letality_rate_1 * 100)
```

Podemos fazer outras estimativas. Se o combined["recovered"] Ã© o nÃºmero de recuperados atÃ© hoje, se eu somar recuperados com mortos tenho todo mundo para quem "acabou a conversa" - ou a pessoa se recuperou ou a pessoa faleceu. Essa pessoa nÃ£o vai aparecer novamente nesse dataset, a nÃ£o ser que o vÃ­rus possa ser contraÃ­do novamente.

```python
letality_rate_2 = combined['deaths'] / (combined['recovered'] + combined['deaths'])
```

```python
sum_up = combined.sum()
letality_rate_1 = sum_up['deaths'] / sum_up['confirmed']
letality_rate_2 = sum_up['deaths'] / (sum_up['recovered'] + sum_up['deaths'])
print(letality_rate_1 * 100)
print(letality_rate_2 * 100)
```

Repare que nÃ£o tem certo ou errado. Em uma estamos considerando todo mundo doente se recuperando e no outro estamos ignorando quem estÃ¡ doente hoje, porque nÃ£o sabemos se vÃ£o se recuperar. Dois cenÃ¡rios totalmente diferentes. Por isso Ã© importante que o nÃºmero resumido venha com uma explicaÃ§Ã£o do motivo de ter chegado nele, as vantagens e desvantagens de utilizar esse nÃºmero.

A vantagem aqui Ã© que sabemos que nÃ£o estamos considerando casos novos. Por outro lado, estamos desconsiderando vÃ¡rios casos que vÃ£o se recuperar. EntÃ£o, sÃ£o dois nÃºmeros.

```python
letality_rate_1 = combined['deaths'] / combined['confirmed'] * 100
combined['letality_rate_1'] = letality_rate_1
combined['letality_rate_2'] = letality_rate_2

combined.head()
```
