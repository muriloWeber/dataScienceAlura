```python
confirmed_by_country = confirmed.groupby('Country/Region').sum()
confirmed_by_country.head()
```

```python
confirmed_by_country.sample(20)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,5))
confirmed_by_country.loc['China'][2:].plot()
# plt.ticklabel_format(style='plain', axis='y')
plt.title('Casos confirmados na China')

plt.show()
```

```python
# exemplo ruim de manusear o eixo y

plt.figure(figsize = (10,5))
confirmed_by_country.loc['China'][2:].plot()
# plt.ticklabel_format(style='plain', axis='y')
plt.ylim(40000, 80000)
plt.title('Casos confirmados na China')

plt.show()
```

```python
new_cases_china = confirmed_by_country.loc['China'][2:]['1/1/22':].diff().dropna()
plt.figure(figsize = (10,5))
plt.title('Casos Novos na China', fontsize = 16)
new_cases_china.plot()
```

```python
plt.figure(figsize = (10,5))
confirmed_by_country[last_date].plot()
plt.title('Confirmados por PaÃ­s')

plt.show()
```

```python
# 99.99% das vezes nÃ£o devemos usar grÃ¡fico de pizza e rosca
# 'se o nome do grÃ¡fico tem comida no meio, nÃ£o use" (boa prÃ¡tica)

# 1. muitas categorias = inviÃ¡vel
# 2. poucas categorias = a comparaÃ§Ã£o serÃ¡ de Ã¡rea

plt.figure(figsize = (10,5))
confirmed_by_country[last_date].plot(kind='pie')
plt.title('Confirmados por PaÃ­s')

plt.show()
```

```python
plt.figure(figsize = (20,5))
confirmed_by_country[last_date].plot(kind='bar')
plt.title('Confirmados por PaÃ­s')

plt.show()
```

```python
plt.figure(figsize = (20,5))
confirmed_by_country[last_date].sort_values(ascending = False)[0:10].plot(kind='bar')
plt.title('Confirmados por PaÃ­s')

plt.show()

# Lembrando que existe um vÃ­cio aqui, pois paises que nÃ£o testaram individuos
# podem sim ter casos ocorrendo, mas nÃ£o confirmados
```
